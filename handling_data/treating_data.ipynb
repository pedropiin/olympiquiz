Como não iremos utilizar o peso nem a altura dos atletas, podemos retirar as colunas do nosso dataframe. Além disso, removeremos as demais colunas que não serão utilizadas:

```python
athletes.drop(['Height', 'Weight', 'Team', 'Games', 'Season', 'City', 'Event'], axis=1, inplace=True)
```

```python
athletes.isna().sum()
```

Agora, para tratar os atletas que não tem a idade definida, vamos apenas removê-los, enquanto os que não ganharam medalha, vamos definir como "No Medal"

```python
athletes.dropna(subset=['Age'], inplace=True)
```

```python
athletes.fillna({'Medal': 'No Medal'}, inplace=True)
```

Excluindo nomes que começam com parenteses (apelidos), números (II, por exemplo), ou vírgulas

```python
single_names = lambda x: (x.split())[0] + " " + (x.split())[-1] if (x.split())[-1][0] not in "(-,IIIV" or len(x) == 2 else (x.split())[0] + " " + (x.split())[-2]
```

```python
last_names = lambda x: (x.split())[0] + " " + (x.split())[-1] if (x.split())[-1][-1] not in "(-,IIIV" or len(x) == 2 else (x.split())[0] + " " + (x.split())[-1][0:-1]
```

```python
sec = athletes.copy()
sec["Name"] = sec["Name"].apply(single_names)
sec["Name"] = sec["Name"].apply(last_names)
sec.sample(30)
```

```python
test = sec["ID"].apply(lambda x: str(x))
test.duplicated().sum()
# Ou seja, temos 132439 itens duplicados, todos de atletas que participaram de mais de uma olimpíada
```

```python
sec.to_csv('athletes_clean.csv')
```
